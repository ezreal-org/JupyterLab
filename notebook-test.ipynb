```python
import matplotlib.pyplot as plt
import numpy as np

# 从 −π−π 到 +π+π 等间隔的 256 个值
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)

plt.plot(X, C)
plt.plot(X, S)
plt.show()
```
